```python
##how to handle with missing values 
```

```python
ufo=pd.read_csv('http://bit.ly/uforeports')
ufo.head()
```

```python
ufo.tail()   ## gives the last five rows     Nan stands for Not a number and it is about missing a value 
```

```python
ufo.isnull().tail()    # True for missing value
```

```python
ufo.notnull().tail()
```

```python
ufo.isnull().sum(axis=0)  # gives the numbers of missing value for each columns   here axis=0
```

```python
pd.Series([True,False,True]).sum()  # it sums the true 
```

```python
ufo[ufo.City.isnull()]
```

```python
ufo.shape    # gives the shape of the DataFrame
```

```python
ufo.dropna(how='any').shape  # drop where rows where at least one value is Nan
```

```python
ufo.dropna(how='all').shape  # drop a row where all the values are NaN
```

```python
ufo.dropna(subset=['City','Shape Reported'],how='any').shape # this is saying a i want to drop all rows where at least City or Shape repo is Nan
```

```python
ufo['Shape Reported'].value_counts()  # how many times each value is repeated in the colomuns Shape Reported
```
